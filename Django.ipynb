**Anotações da aula de introdução DJANGO e conceitos fundamentais para iniciantes**


      *CONCEITOS BÁSICOS*

**1. CLIENTE:**
É quem faz o pedido

**2. SERVIDOR:**
É quem processa o pedido e devolve a resposta

**COMO FUNCIONA?**
1. Cliente faz uma requisição.
2. Servidor recebe.
3. Servidor processa.
4. Servidor devolve uma resposta.

**HTTP:**

* Conjunto de regras que define um cliente(navegador), envia um pedido e como o servidor responde.

* REQUISIÇÃO:
1. Método HTTP(Get, Post...)
2. URL (Endereço do recurso)
3. HEADERS ( Informações adicionais)
4. BODY (Corpo- dados enviados, geralmente em POST)

* RESPOSTA HTTP:
1. Código de status;
2. Headers;
3. Body( Normalmente HTML ou JSON )
