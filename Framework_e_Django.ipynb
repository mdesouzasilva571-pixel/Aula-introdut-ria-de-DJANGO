**O QUE É DJANGO?**

* Framework web em Python, criado para facilitar o desenvolvimento de aplicações escaláveis e seguras.

* Oferece:

1. Servidor interno web;
2. Sistemas de rotas (URLs);
3. Conexão com banco de dados;
4. ORM (Mapear objtos para tabela);
5. Painel administrativo automático;
6. Segurança embutida;
7. Templates(HTML integrado);
8. Autenticação de usuários;

* Por que aprender?

1. Rápido para desenvolver;
2. Muito seguro;
3. Usuado por empresas grandes;
4. Código organizado;
5. Comunidade enorme;
